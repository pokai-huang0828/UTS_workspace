## 1) 环境与数据（Environment & Data）
运行下面的单元以导入库并载入示例数据 **Anscombe**。

```python
import seaborn as sns
import pandas as pd

# 载入示例数据：Anscombe（包含四组经典测试数据集 I~IV）
anscombe = sns.load_dataset("anscombe")
anscombe.head()
```

## 2) 基础回归散点图（四组数据同图）
去掉置信区间（`ci=None`）以更清晰地观察差异；使用 `hue='dataset'` 区分子数据集。

```python
sns.lmplot(data=anscombe, x="x", y="y", hue="dataset", ci=None)
```

## 3) 分面显示（按子数据集分离）
每个子图展示一个子数据集，便于比较不同分布特征。

```python
sns.lmplot(data=anscombe, x="x", y="y", col="dataset", hue="dataset", ci=None)

sns.lmplot(data=anscombe, x="x", y="y", col="dataset", hue="dataset")
```

## 4) 布局优化（2×2 网格 with `col_wrap`）
当屏幕较小时，可使用 `col_wrap=2` 将分面对齐为 2×2 网格。

```python
sns.lmplot(data=anscombe, x="x", y="y", col="dataset", hue="dataset", ci=None, col_wrap=2)
```

## 5) 加上置信区间（Confidence Interval）
去掉 `ci=None` 即会显示回归线的置信带，以观察不确定性。

```python
sns.lmplot(data=anscombe, x="x", y="y", col="dataset", hue="dataset")  # 默认显示置信区间
```

## 6) 多项式回归（Polynomial Regression）
对 Series II 进行二次与三次拟合，比较非线性关系的拟合效果。

```python
# 二次拟合
sns.lmplot(
    data=anscombe.query("dataset == 'II'"),
    x="x", y="y", ci=None, order=2
)
```

```python
# 三次拟合
sns.lmplot(
    data=anscombe.query("dataset == 'II'"),
    x="x", y="y", ci=None, order=3
)
```
